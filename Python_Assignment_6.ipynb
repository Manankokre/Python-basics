2.  How does broadcasting work in NumPy?
  - Broadcasting in NumPy is a powerful mechanism that allows operations to be performed on arrays of different shapes without needing to copy data or explicitly reshape them. When two arrays are operated together, NumPy compares their shapes starting from the trailing (rightmost) dimensions and attempts to match them. If the dimensions are equal or one of them is 1, NumPy can “stretch” the smaller array across the larger one to make their shapes compatible. This automatic expansion happens virtually, meaning no extra memory is used to actually duplicate values. Broadcasting makes vectorized operations possible, allowing you to apply arithmetic operations to entire arrays element-wise, even if their shapes differ. It greatly simplifies code and improves performance because it eliminates the need for explicit loops and manual repetition of data.

8. What does the term “vectorized operation” mean in NumPy?
  - In NumPy, the term **“vectorized operation”** refers to performing operations on entire arrays or large sets of data **at once**, without writing explicit loops in Python. Instead of processing elements one by one, NumPy uses optimized low-level C code to apply operations (like addition, multiplication, or comparisons) to whole arrays simultaneously. This makes vectorized operations much faster and more efficient than traditional looping in Python. For example, adding two arrays with `a + b` in NumPy automatically applies the addition to each element, whereas with lists you would need a loop. So, vectorization improves performance, reduces code complexity, and allows numerical computations to be done in a clean and highly efficient manner.

17.  Explain the difference between apply() and map() in Pandas.
  - In Pandas, both `apply()` and `map()` are used to apply a function to data, but they differ in their scope and use cases.

    - **`map()`** is primarily used on **Series** objects to apply a function element-wise. It can also accept a dictionary or a mapping of values to transform data. For example, it’s ideal for converting, replacing, or transforming individual values in a column.

    - **`apply()`** is more flexible and can be used on both **Series and DataFrames**. On a Series, it works similarly to `map()`, applying a function element-wise. On a DataFrame, however, it can operate **row-wise or column-wise**, allowing you to perform more complex operations that involve multiple columns at once.
18. What are some advanced features of NumPy?
  - NumPy offers several advanced features that make it a powerful library for numerical computing beyond basic array operations:

    1. **Broadcasting** - Allows arithmetic operations on arrays of different shapes, automatically expanding them to be compatible without explicit loops. This enables efficient element-wise computations.

    2. **Vectorized Operations** - Supports operations on entire arrays without Python loops, which leads to faster computation and cleaner code.

    3. **Universal Functions (ufuncs)** - Provides optimized, element-wise functions for mathematical, logical, and statistical operations, such as `np.sin()`, `np.exp()`, or `np.sqrt()`.

    6. **Random Number Generation** - Advanced random sampling, distributions, and reproducible randomness through `numpy.random` module.

    7. **Structured Arrays** - Support for arrays with complex, heterogeneous data types, similar to a lightweight DataFrame.

    8. **Masking and Boolean Indexing** - Allows filtering and conditional selection of data efficiently using boolean arrays.

    9. **Fourier Transforms and Signal Processing** - Functions like `np.fft` enable efficient frequency domain analysis.

    10. **Integration with Other Libraries** - NumPy arrays serve as the backbone for libraries like Pandas, SciPy, Matplotlib, TensorFlow, and PyTorch, making it central to scientific computing and machine learning in Python.

21. Why is NumPy's array slicing faster than Python's list slicing?
  - NumPy’s array slicing is faster than Python’s list slicing primarily because of **how the data is stored and accessed in memory**. NumPy arrays store elements in **contiguous blocks of memory** and require all elements to be of the same data type. This allows slicing operations to create **views** of the original array instead of copying the data, meaning no new memory allocation is needed and the operation is performed at a low level in optimized C code. In contrast, Python lists store references to objects scattered in memory, so slicing a list often requires creating a **new list and copying references**, which is slower. Additionally, NumPy’s internal implementation is **vectorized**, leveraging highly optimized algorithms, whereas Python’s list operations rely on slower interpreted loops. As a result, NumPy slicing is both **memory-efficient and significantly faster**, especially for large datasets.

4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample DataFrame
data = {
    'A': [1, 2, 3, 4, 5],
    'B': [5, 4, 3, 2, 1],
    'C': [2, 3, 4, 5, 6],
    'D': [5, 3, 6, 2, 1]
}

df = pd.DataFrame(data)

# Calculate the correlation matrix
corr_matrix = df.corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

11. Perform matrix multiplication using NumPy.

```python
import numpy as np

# Create two matrices
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

# Perform matrix multiplication
C = np.dot(A, B)  # Method 1
# Alternatively, you can use the @ operator: C = A @ B

print("Matrix A:")
print(A)
print("Matrix B:")
print(B)
print("Matrix Multiplication Result (A x B):")
print(C)
```
