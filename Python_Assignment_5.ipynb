14.  What is the role of try and except in exception handling?
  - In Python’s exception handling, the **`try`** and **`except`** blocks work together to prevent a program from crashing when an error occurs. The **`try` block** contains the code that might cause an exception—Python attempts to run this code normally. If everything works correctly, the program continues without interruption. However, if an error happens inside the `try` block, Python immediately stops executing that block and looks for a matching **`except` block**. The **`except` block** defines how the program should respond to that specific error, such as showing a user-friendly message or performing another safe action. This structure helps manage errors gracefully, ensuring the program continues to run instead of terminating unexpectedly.
15.  How does Python's garbage collection system work?
  - Python’s garbage collection system automatically manages memory by identifying and removing objects that are no longer needed. The primary mechanism it uses is **reference counting**, where every object keeps track of how many references point to it. When this count drops to zero—meaning the program can no longer access the object—Python immediately frees the memory. However, reference counting alone cannot handle **circular references** (when two objects refer to each other), so Python also includes a **cyclic garbage collector**. This garbage collector periodically scans groups of objects to detect cycles that are no longer reachable from the program. If such cycles are found, Python safely deletes them, freeing memory. Overall, the garbage collection system ensures efficient memory management without requiring the programmer to manually allocate or free memory.
16.  What is the purpose of the else block in exception handling?
  - In Python’s exception handling, the **`else` block** is used to define code that should run **only if no exception occurs** in the `try` block. It separates normal execution from error-handling logic, making the program cleaner and easier to understand. When the code inside `try` runs successfully without raising any exception, Python skips all the `except` blocks and executes the `else` block. This is useful for actions that should happen only when everything goes right, such as processing results, displaying success messages, or performing operations that depend on the successful completion of the `try` block. By keeping this code in an `else` block rather than inside `try`, you avoid accidentally catching errors that do not belong there, leading to clearer and more predictable behavior.

18.  What is the difference between os.fork() and multiprocessing in Python?
  - **os.fork()** is a low-level Unix-specific system call that directly creates a new child process by duplicating the current process. It is available only on Unix-like systems (Linux, macOS) and not on Windows. With `os.fork()`, the child process receives a copy of the parent’s memory space, and both continue executing from the point where the fork happened. Because it is low-level, it requires manual handling of process communication, synchronization, and cleanup. It is powerful but less portable and harder to manage for complex applications.

  - **multiprocessing**: It is a high-level Python module designed to create and manage processes in a platform-independent way. It works on Windows, macOS, and Linux and provides features like process pools, queues, pipes, and shared memory for easier inter-process communication. The `multiprocessing` module abstracts the complexity of low-level forking, making it easier to write parallel programs that run safely across different operating systems. It is more user-friendly, structured, and suitable for most real-world multiprocessing tasks.

23. What are the challenges associated with memory management in Python?
  - Memory management in Python comes with several challenges despite its built-in garbage collector. One challenge is **memory leaks**, which can occur when objects remain referenced unintentionally, preventing the garbage collector from freeing them. Another issue is **circular references**, where objects reference each other, making them harder for the garbage collector to clean up efficiently. Python also struggles with **high memory usage** because objects often consume more memory compared to lower-level languages, especially when handling large datasets. Additionally, the **Global Interpreter Lock (GIL)** can limit memory-related performance in multi-threaded programs, preventing true parallel execution. Managing memory in long-running applications becomes challenging because developers must ensure efficient object usage, proper cleanup, and avoid unnecessary data retention. Overall, while Python simplifies memory management, developers still need to be mindful of how objects are created, referenced, and released.

```python
#4. Write a Python script that reads from one file and writes its content to another file.
# Read from source file and write to destination file

# Source file: input.txt
# Destination file: output.txt

try:
    with open("input.txt", "r") as src:
        with open("output.txt", "w") as dest:
            for line in src:
                dest.write(line)
    print("File copied successfully!")
except FileNotFoundError:
    print("Error: Source file does not exist.")
```

```python
#6. Write a Python program that logs an error message to a log file when a division by zero exception occurs.
import logging

# Configure logging
logging.basicConfig(
    filename="errors.log",     # Log file name
    level=logging.ERROR,       # Log only error messages or above
    format="%(asctime)s - %(levelname)s - %(message)s"
)

try:
    a = 10
    b = 0
    result = a / b
    print(result)

except ZeroDivisionError:
    logging.error("Division by zero error occurred.")
    print("An error occurred. Check the log file for details.")
```

```python
#18. How would you implement a basic logging setup that logs to a file with rotation after 1MB.
import logging
from logging.handlers import RotatingFileHandler

# Create a rotating file handler
handler = RotatingFileHandler(
    "app.log",        # Log file name
    maxBytes=1_000_000,  # Rotate after 1 MB
    backupCount=5        # Keep 5 backup log files
)

# Configure the logging format
formatter = logging.Formatter("%(asctime)s - %(levelname)s - %(message)s")
handler.setFormatter(formatter)

# Set up the logger
logger = logging.getLogger()
logger.setLevel(logging.INFO)
logger.addHandler(handler)

# Example log messages
logger.info("Application started.")
logger.error("This is an error message.")
logger.warning("This is a warning message.")
```

```python
#21. Write a Python program that reads a file and prints the number of occurrences of a specific word.
# Specify the word to search for
word_to_find = "python"

count = 0

try:
    with open("example.txt", "r") as file:
        for line in file:
            # Convert to lowercase to make search case-insensitive
            words = line.lower().split()
            count += words.count(word_to_find.lower())

    print(f"The word '{word_to_find}' appears {count} times in the file.")

except FileNotFoundError:
    print("Error: The file does not exist.")
```

```python
#23. Write a Python program that writes to a log file when an error occurs during file handling.
import logging

# Set up logging configuration
logging.basicConfig(
    filename="file_errors.log",      # Log file
    level=logging.ERROR,             # Only log errors
    format="%(asctime)s - %(levelname)s - %(message)s"
)

try:
    # Attempt to open and read a file
    with open("data.txt", "r") as file:
        content = file.read()
        print(content)

except FileNotFoundError:
    logging.error("FileNotFoundError: The file 'data.txt' was not found.")
    print("Error: File not found. Check log for details.")

except PermissionError:
    logging.error("PermissionError: Permission denied while accessing 'data.txt'.")
    print("Error: Permission denied. Check log for details.")

except Exception as e:
    logging.error(f"Unexpected error: {e}")
    print("An unexpected error occurred. Check log for details.")
```
